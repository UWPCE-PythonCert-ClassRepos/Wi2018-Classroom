```python
"""
Using the same data set, write a closure to capture high energy tracks. 
We will define high energy tracks as anything over 8.0. Submit your code 
and the tracks it finds, artist name, track name and energy value.
"""

def high_energy_closure(dataframe):
    def find_high_energy(minimum_level):
        # nonlocal dataframe  -- hmmm, this also works when commented out. Because it's mutable?
        return ((song_info[0], song_info[1], song_info[2]) for song_info in zip(dataframe.energy, dataframe.artists, dataframe.name) if song_info[0] > minimum_level)
    return find_high_energy
```

```python
find_high = high_energy_closure(music)
print(*find_high(.80)) # Energy is out of "1.0" so I'm using .8 instead of "8.0"
```

```python
for tup in find_high(.80):
    print(f" {tup[2]}, By: {tup[1]}, Has energy of: {tup[0]:.3f}" )
```
